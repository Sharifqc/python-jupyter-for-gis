```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_map():

    fig = plt.figure(figsize=(6,6))
    
    map = Basemap(projection='spaeqd',lon_0=0,resolution='l', boundinglat=-60)

    map.bluemarble()

    map.drawparallels(np.arange(-90.,-45,10.))
    map.drawmeridians(np.arange(0.,360.,60.))
    
    return fig
```

```python
f = draw_map()
```

```python
f.savefig('figures/2.png', bbox_inches='tight')
```
